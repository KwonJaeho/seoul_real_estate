```python
# 원하는 건물 입력시 가격 변동 확인 (그래프)
word = input("건물명: ")
build = data.groupby('도로명주소').get_group(word)
build=build.reset_index(drop=True)
```

```python
#그래프 표현
x = build['접수연도']
y = build['물건금액(만원)']
plt.plot(x,y,marker='o',linestyle='--')
plt.xlabel("접수연도")
plt.ylabel("물건금액(만원)")
plt.xticks(x)
plt.yticks(y)
plt.title(word)
plt.show()
```

```python
# max = 최대값  min = 최솟값 std = 표주편차
test = data.groupby('도로명주소').filter(lambda x: x['도로명주소'].count() > 1)  # 거래가 2건이상의 건물만 필터링
build_list = test.groupby('도로명주소').agg({'도로명주소':'count','물건금액(만원)':['min','max']}).reset_index() # 최고 가격 및 최소 가격
res = []
for i in range(len(build_list)):
    min = build_list['물건금액(만원)','min'].loc[i]
    max = build_list['물건금액(만원)','max'].loc[i]
    res.append(round(((max - min)/min)*100))
```

```python
build_list['res'] = res
```

```python
#표현
max = build_list['res'].idxmax()
a = build_list.loc[max]
pd.DataFrame(a)
```
